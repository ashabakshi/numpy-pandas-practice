# concat() — Combine DataFrames by stacking

```python
# concat() is used to combine multiple DataFrames.
# It can stack them vertically (row-wise) or horizontally (column-wise)

#         pd.concat([df1, df2], axis=0)

#         axis=0 → stack rows (default)

#         axis=1 → stack columns
```

```python
import pandas as pd
```

```python
# Vertical stacking (row wise)
df1 = pd.DataFrame({
    'ID':[1,2],
    'Name':['Asha','Priya']
})
print(df1)
df2= pd.DataFrame({
    'ID':[3,4],
    'Name':['Komal','Jahanvi']
})
print(df2)
combined = pd.concat([df1,df2], ignore_index=True)
print(combined)
```

```python
# Horizontal stacking( column wise)
df1=pd.DataFrame({
    'Name':['Asha','Priya']
})
print(df1)
df2= pd.DataFrame({
    'Marks':[90,85]
})
print(df2)
combined = pd.concat([df1,df2], axis=1)
print(combined)
```

# merge() — Like SQL JOIN

```python
# merge() is used to combine two DataFrames based on a common column (key).

# It works just like SQL joins (e.g., inner, left, right, outer joins).
```

```python
# syntax
# pd.merge(left_df, right_df, on='key_column', how='type_of_join')
# on → the column name to join on

# how → type of join: 'inner', 'left', 'right', 'outer'
```

```python
# merge() (Inner Join
students= pd.DataFrame({
    'ID':[1,2,3,4],
    'Name':['Asha','Komal','Jahanvi','Prateek']
})
print(students)
marks=pd.DataFrame({
    'ID':[1,2,3],
    'Marks':[90,85,88]
})
print(marks)
merged= pd.merge(students,marks,on='ID')
print(merged)
```

```python
# Example — Left Join
left_joined = pd.merge(students, marks,on='ID', how='left')
print(left_joined)
```

#Example — Outer Join
outer_joined = pd.merge(students, marks,on='ID', how='outer')
print(outer_joined)

```python
#Example — inner Join
inner_joined = pd.merge(students, marks,on='ID', how='inner')
print(inner_joined)
```

```python
#Example — right Join
right_joined = pd.merge(students, marks,on='ID', how='right')
print(right_joined)
```

## Grouping and aggregating (gruopby)


```python
# Syntax
# df.groupby('column_name').agg_function()
# df.groupby('column_name')['target_column'].agg_function()
```

```python
import pandas as pd

data=pd.DataFrame({
    'Department':['IT','HR','IT','HR','IT'],
    'Salary':[50000,60000,52000,61000,51000]
})
grouped= data.groupby('Department')['Salary'].sum()
print(grouped) #It grouped by Department and summed the Salary.
```

# Multiple Aggregations

```python
print(data)
data.groupby('Department')['Salary'].agg(['mean','max','count'])
```

# Group by multiple columns

```python
data=pd.DataFrame({
    'Department':['IT','HR','IT','HR','IT'],
    'Gender':['F','F','M','M','F'],
    'Salary':[50000,60000,52000,61000,51000]
})
result=data.groupby(['Department','Gender'])['Salary'].mean()
print(result)
```

```python
import pandas as pd
df= pd.DataFrame({
    'Name':['Asha','Riya','Neha'],
    'Marks':[85,92,78]
})
sorted_df= df.sort_values(by='Marks')
print(sorted_df)
```

# Sort by multiple columns




```python
df= pd.DataFrame({
    'Name':['Asha','Riya','Neha','Asha'],
    'Marks':[85,92,85,92]
})
df.sort_values(by=['Marks','Name'],ascending=[False,True])
```
