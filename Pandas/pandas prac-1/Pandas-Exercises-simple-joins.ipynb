### Step 6. concat cars1 and cars2 into a single DataFrame called cars

```python
def con(cars1,cars2):
    if sorted(cars1.columns)==sorted(cars2.columns):
        return pd.concat([cars1,cars2],axis=0).reset_index(drop=True)
    else:
        return pd.concat([cars1,cars2],axis=1).reset_index(drop=True)
```

```python
cars= con(cars1,cars2)
cars
```
