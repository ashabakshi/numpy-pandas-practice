# propties of array
- ndim
- shape
- size
- dtype

```python
a=np.array([1,2,3,4,5])
print(a)
print(type(a))
print(a.ndim)
print(a.shape)
```

```python
b=np.array([[1,2,3,4,5]])
print(b,type(b))
print(b.ndim)
print(b.shape)
print(b.size)
```

```python
c=np.array([[1,2,3],[5,6,7],[7,8,9]])
print(c,type(c))
print(c.ndim)
print(c.shape)
print(c.size)
print(c.dtype)
```

```python
x=np.array([[1],[3],[4]])
print(x)
print(x.ndim)
print(x.shape)
print(x.size)
```

```python
d=np.array([[[[1,2,3],[5,6,7],[7,8,9]]]])
print(d,type(d))
print(d.ndim)
print(d.shape)
print(d.size)
```

```python
e=np.array([[[[1,2,3],[5,6,72],[7.1,8,9],[7,8,9],[1,1,1]]]])
print(e,type(e))
print(e.ndim)
print(e.shape)
print(e.size)
print(e.dtype)
```
