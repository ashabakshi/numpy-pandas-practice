```python
L=['stone','paper','scissor']
u_score = 0
c_score=0
chance = int(input('enter chance '))
while chance !=0:
    
    comp = np.random.choice(L)
    user = input('play the game ')
    
    
    if comp==user:
        print('Tie')
    
    elif comp=='stone':
        if user=='paper':
            print('user wins')
            u_score +=1

        else:
            print('comp wins')
            c_score +=1

    
    elif comp=='paper':
        if user=='scissor':
            print('user wins')
            u_score +=1

        else:
            print('comp wins')
            c_score +=1

    
    else:
        if user=='stone':
            print('user wins')
            u_score +=1

        else:
            print('comp wins')
            c_score +=1

    chance-=1


if c_score==u_score:
    print('No wins')

elif c_score>u_score:
    print('Comp wins by',c_score)
else:
    print('user wins by',u_score)




        
        
```
